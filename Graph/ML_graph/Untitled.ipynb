```python
# Import libraries
import markov_clustering as mc
import networkx as nx

# Load network data
G = nx.karate_club_graph()

# Build adjacency matrix
A = nx.to_numpy_matrix(G)
```

```python
# Run MCL algorithm
result = mc.run_mcl(A)
clusters = mc.get_clusters(result)

# Draw clusters
mc.draw_graph(A, clusters, node_size=300, with_labels=True, edge_color="silver")
```

```python
G = nx.path_graph(8)
A = nx.to_numpy_matrix(G)
result = mc.run_mcl(A)
clusters = mc.get_clusters(result)

# Draw clusters
mc.draw_graph(A, clusters, node_size=300, with_labels=True, edge_color="silver")
```

```python
G = nx.DiGraph()
G.add_edges_from([(0, 1), (0, 2), (0, 3), (0, 4)])
A = nx.to_numpy_matrix(G)
result = mc.run_mcl(A)
clusters = mc.get_clusters(result)

# Draw clusters
mc.draw_graph(A, clusters, node_size=300, with_labels=True, edge_color="silver")
```
